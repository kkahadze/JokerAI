```python
from joker_simulation import Simulation
import sys
import pandas as pd
import pickle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
sim = Simulation()
sim.run(1000, False)
```

```python
df = pd.DataFrame()
df['place'], df['dealt'] ,df['wild'] = (sim.places), (sim.dealt), (sim.wild)
df['num_wild'], df['num_joks'], df['already_called'] = sim.num_wild, sim.num_joks, sim.already_called
df['taken'] = sim.taken
df
```

```python
df['dealt'] = (df.dealt - 1) / 8
df['wild'] = (df.wild).astype(int)
df['num_wild'] = df.num_wild / 9
df['num_joks'] = df.num_joks / 2
df['already_called'] = df.already_called / 27
df['taken'] = df.taken
df['place'] = (df.place - 1) / 4
df
```

```python
df.dealt.describe()
```

```python
X = np.asarray(df.iloc[:, :6]).reshape(-1 , 6)
y = np.asarray(df.iloc[:, 6])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .1, random_state = 0)
```

```python
clf = KNeighborsClassifier(n_neighbors= 13)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
scores = []
for i in range(1, 50):
    clf = KNeighborsClassifier(n_neighbors= i)
    clf.fit(X_train, y_train)
    scores.append(str(i) + " " + str(clf.score(X_test, y_test)))
# for score in scores: print(score)
```
